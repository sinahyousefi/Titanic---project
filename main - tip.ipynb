```python
tip['bucket'] = np.where(tip['total_bill'] < 20 , 'low', 'high')
tip.head()
```

```python
conditions = [
    tip['total_bill'] < 20,                     # bucket 1
    (tip['total_bill'] >= 20) & (tip['total_bill'] < 40),  # bucket 2
    tip['total_bill'] >= 40                     # bucket 3
]

choices = ['low', 'medium', 'high']

tip['bucket'] = np.select(conditions, choices)

tip.head()
```
