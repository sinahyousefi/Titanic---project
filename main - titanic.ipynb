```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder, TargetEncoder
import category_encoders as ce

encoder1 = OneHotEncoder(sparse_output=False)
encoder2 = LabelEncoder()
encoder3 = OrdinalEncoder()
encoder4 = ce.TargetEncoder()

data_encoder2 = data.copy()
data_encoder2['Sex_ord'] = encoder3.fit_transform(data[['Sex']])
data_encoder2['Sex_lab'] = encoder2.fit_transform(data['Sex'])
data_encoder2['Sex_tar'] = encoder4.fit_transform(data['Sex'], data['Fare'])
data_encoder2.head()
```

```python
df = data.copy()
enc_lab = LabelEncoder()

df['Embarked'] = enc_lab.fit_transform(df['Embarked'])
df['Sex'] = enc_lab.fit_transform(df['Sex'])
df.head()
```

```python
df.info()
```

```python
df_age_null = df[df['Age'].isnull()]
df_age_notnull = df[~df['Age'].isnull()]

feature_reg = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Fare', 'Embarked']

x_train = df_age_notnull[feature_reg]
y_train = df_age_notnull['Age']
x_test = df_age_null[feature_reg]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train,y_train)

df_age_null['Age'] = (model.predict(x_test))

df_age_null.head()
```

```python
df.loc[df['Age'].isnull(),'Age'] = model.predict(x_test).round().astype(int)
df.head(10)
```

```python
df_cabin_null = df[df['Cabin'].isnull()]
df_cabin_notnull = df[~df['Cabin'].isnull()]

feature_class = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

x_train_cabin = df_cabin_notnull[feature_class]
y_train_cabin = df_cabin_notnull['Cabin']

x_test_cabin = df_cabin_null[feature_class]
```

```python
from sklearn.ensemble import RandomForestClassifier

model_cabin = RandomForestClassifier()
model_cabin.fit(x_train_cabin, y_train_cabin)


df_cabin_null['Cabin'] = model_cabin.predict(x_test_cabin)
df_cabin_null.head()
```
